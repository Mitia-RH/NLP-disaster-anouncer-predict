```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
df.describe()
```

```python
df.head(10)
```

```python
##### PROCESSING A TEXT TO KEYWORDS #####
from sklearn.feature_extraction.text import CountVectorizer
NotNecesaryWord = [" and ", " are ", " or ", " the ", " is ", " this ", " of ", "  "]
punctuations = [",",".",";","!","?", "\"", "(", ")", "@", "'"]

matrix = []

for text in df.text:
    text = text.lower()
    text = " " + text + " "
    for w in NotNecesaryWord: 
        text = text.replace(w, ' ')
    for p in punctuations:
        text = text.replace(p, '')
    matrix.append(text)

vectorText = CountVectorizer()
XVec = vectorText.fit_transform(matrix)
feature_names= vectorText.get_feature_names_out()
X_array = XVec.toarray()

df['vector'] = [vector for vector in X_array]
feature_model = ["keyword", "location", "target"]
Y = df["target"]
X = X_array
```

```python
df.head()
```

```python
print('Feature name size : ',feature_names.size)
print('A vector dimension size : ', X_array[0].size)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

rs = 42
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, Y, random_state=rs)
```

```python
classifier = LogisticRegression(random_state=rs)
lr_model = classifier.fit(Xtrain, Ytrain)
lr_model_ac = accuracy_score(lr_model.predict(Xtest), Ytest)
```

```python
print(f'Accuracy of the model : {lr_model_ac}')
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

df = pd.read_csv('train.csv')
vectorizer = CountVectorizer()
X_text_vectorized = vectorizer.fit_transform(df['text'])

other_features = pd.get_dummies(df[['keyword', 'location']])


# Prepare target labels
X = pd.concat([pd.DataFrame(X_text_vectorized.toarray()), other_features], axis=1)
X.columns = X.columns.astype(str)
Y = df['target']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=rs)

classifier = LogisticRegression(random_state=rs)
classifier.fit(X_train, Y_train)

Y_pred = classifier.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred)

print("Accuracy with keyword and location :", accuracy)
```
